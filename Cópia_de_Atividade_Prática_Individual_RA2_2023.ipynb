# Questão 1 (vale 6 pontos)
Considerando as funções listadas a seguir faça: **(a)** o gráfico das três funções; **(b)** o passo a passo algébrico, em latex, para encontrar o ângulo de inclinação da reta tangente no ponto $x=5$ para cada das funções:

1. $
f(x) = (x^2 - 4)(x^3 - 7x + 1)
$
2. $
f(x) = \sin(x^2)\cos(2x)
$
3.$
f(x) = x^2e^{-x}\ln(x)
$

Se você usar o sympy, terá que fazer com que ele imprima todos os passos necessários para resolver cada uma das operações, regras, ou tabelas que aplicar para resolver o problema.

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio.


```python
# (a)
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f1 = (x**2 - 4)*(x**3 - 7*x + 1)
f2 = sp.sin(x**2)*sp.cos(2*x)
f3 = x**2*sp.exp(-x*sp.ln(x))

f1_np = sp.lambdify(x, f1, 'numpy')
f2_np = sp.lambdify(x, f2, 'numpy')
f3_np = sp.lambdify(x, f3, 'numpy')

x_range = np.linspace(-5, 5, 500)

plt.figure(figsize=(10, 4))

# 1
plt.subplot(1, 3, 1)
plt.plot(x_range, f1_np(x_range))
plt.title(r'$f(x)=(x^2-4)(x^3-7x+1)$')

# 2
plt.subplot(1, 3, 2)
plt.plot(x_range, f2_np(x_range))
plt.title(r'$f(x)=\sin(x^2)\cos(2x)$')

# 3
plt.subplot(1, 3, 3)
plt.plot(x_range, f3_np(x_range))
plt.title(r'$f(x)=x^2e^{-x\ln(x)}$')

plt.tight_layout()
plt.show()

# (b)
from sympy import *
from IPython.display import display, Math

df1 = diff(f1)
df2 = diff(f2)
df3 = diff(f3)

m1 = df1.subs(x, 5)
m2 = df2.subs(x, 5)
m3 = df3.subs(x, 5)

theta1 = atan(m1)
theta2 = atan(m2)
theta3 = atan(m3)

# Imprimindo

# 1)
print("\n")
display(Math(r"\text{Passo a passo para } f(x) = (x^2 - 4)(x^3 - 7x + 1):"))
print("\n")
display(Math(r"f(x) = " + latex(f1)))
display(Math(r"\frac{df}{dx} = " + latex(df1)))
display(Math(r"m = " + latex(m1)))
display(Math(r"\theta = " + latex(theta1)))
display(Math(r"\text{Resultado final:}"))
display(Math(r"m = " + latex(m1.evalf())))
display(Math(r"\theta = " + latex(theta1.evalf()) + r"\text{ radianos ou }" + latex(theta1.evalf()*180/pi) + r"\text{ graus}"))
print("\n\n--------------\n\n")

# 2)
display(Math(r"\text{Passo a passo para } f(x) = \sin(x^2)\cos(2x):"))
print("\n")
display(Math(r"f(x) = " + latex(f2)))
display(Math(r"\frac{df}{dx} = " + latex(df2)))
display(Math(r"m = " + latex(m2)))
display(Math(r"\theta = " + latex(theta2)))
display(Math(r"\text{Resultado final:}"))
display(Math(r"m = " + latex(m2.evalf())))
display(Math(r"\theta = " + latex(theta2.evalf()) + r"\text{ radianos ou }" + latex(theta2.evalf()*180/pi) + r"\text{ graus}"))
print("\n\n--------------\n\n")


# 3)
display(Math(r"\text{Passo a passo para } f(x) = x^2e^{-x\ln(x)}:"))
print("\n")
display(Math(r"f(x) = " + latex(f3)))
display(Math(r"\frac{df}{dx} = " + latex(df3)))
display(Math(r"m = " + latex(m3)))
display(Math(r"\theta = " + latex(theta3)))
display(Math(r"\text{Resultado final:}"))
display(Math(r"m = " + latex(m3.evalf())))
display(Math(r"\theta = " + latex(theta3.evalf()) + r"\text{ radianos ou }" + latex(theta3.evalf()*180/pi) + r"\text{ graus}"))




```

Uma empresa espacial está tentando descobrir o ponto máximo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=x^4−5x^3+10x^2−10x+5$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
x = symbols('x')
f = x**4 - 5*x**3 + 10*x**2 - 10*x + 5

f_prime = f.diff(x)
f_double_prime = f_prime.diff(x)

crit_points = solve(f_prime, x)
crit_points = [N(nsimplify(p)) for p in crit_points if p.is_real]
crit_points = [p for p in crit_points if -10 <= p <= 10]

min_val = None

for p in crit_points:
    val = f.subs(x, p)
    if min_val is None or val < min_val:
        min_val = val
        min_point = p

print("Primeiro ponto de mínimo do intervalo de [-10, 10]: ", round(min_point, 4))

# Gráfico
x_vals = np.linspace(-10, 10, 1000)
y_vals = np.array([f.subs(x, xi) for xi in x_vals])

plt.plot(x_vals, y_vals)
plt.scatter(min_point, f.subs(x, min_point), color='green')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import plotly.graph_objs as go

def f(x,y):
    return np.sin(x*y) + np.cos(x+y)

x = np.linspace(0,5,100)
y = np.linspace(0,5,100)

z = np.zeros((100,100))
for i in range(100):
    for j in range(100):
        z[i,j] = f(x[i],y[j])


max_point = np.unravel_index(np.argmax(z), z.shape)
min_point = np.unravel_index(np.argmin(z), z.shape)

# Cria o gráfico
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])
fig.add_trace(go.Scatter3d(x=[x[max_point[0]]], y=[y[max_point[1]]], z=[z[max_point]], mode='markers', marker=dict(size=10, color='red')))
fig.add_trace(go.Scatter3d(x=[x[min_point[0]]], y=[y[min_point[1]]], z=[z[min_point]], mode='markers', marker=dict(size=10, color='green')))
fig.update_layout(title='Gráfico de f(x,y)', autosize=False,
                  width=700, height=700,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```
