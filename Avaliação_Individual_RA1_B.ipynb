# Questão 1: 
O valor de um computador deprecia exponencialmente com o tempo. Seu valor inicial é de R$ 4.000, e a taxa de depreciação anual é de 20%. Qual será o valor do computador após 5 anos? Use a fórmula:

$$
V(t) = V_0 \cdot (1 - r)^t
$$

onde $V_0$ é o valor inicial, $r$ é a taxa de depreciação e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o o valor do bem depreciado, ano a ano, nos próximos 15 anos.

```python
import matplotlib.pyplot as plt

valorInicial = 4000
taxaDep = 0.2
tempo = 5
def formula(p_valorInicial,p_taxaDep,p_tempo):
  return p_valorInicial * (1 - p_taxaDep) ** p_tempo

print("Valor após 5 anos:",round(formula(valorInicial,taxaDep,tempo),2))
```

```python
anos = range(1,16)
valoresPorAno = [ valorInicial * (1 - taxaDep) ** ano for ano in anos ]

plt.plot(anos, valoresPorAno, marker='o')
plt.title('Depreciação anual do valor do computador')
plt.xlabel('Anos')
plt.ylabel('Valor atualizado (R$)')

print("Gráfico:")
plt.show()
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $3^{x-1} - 2^x = 7$

2. $\log_5(2x + 1) + \log_5(x - 2) = 2$

```python
# 1)
import numpy as np

def f1(x):
    return 3**(x-1) - 2**x - 7

x = np.linspace(-1, 1, 500)
y = f1(x)

fig, ax = plt.subplots()
ax.plot(x, y, label=r'$3^{x-1} - 2^x = 7$', color='red')

if np.all(np.diff(y) > 0):
    ax.set_xlabel('Função crescente')
else:
    ax.set_xlabel('Função decrescente')

ax.legend()
plt.show()

```

```python
# 2)

def f2(x):
    return np.log(2*x+1) / np.log(5) + np.log(x-2) / np.log(5) - 2
    

x = np.linspace(2, 10, 500)
y = f2(x)

fig, ax = plt.subplots()
ax.plot(x, y, label=r'$\log_5 (2x+1) + \log_5 (x-2) = 2$', color='blue')

if np.all(np.diff(y) > 0):
    ax.set_xlabel('Função crescente')
else:
    ax.set_xlabel('Função decrescente')

ax.legend()
plt.show()
```
