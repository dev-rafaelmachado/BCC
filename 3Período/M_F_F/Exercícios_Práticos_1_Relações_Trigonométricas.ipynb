1. Um fotógrafo está posicionado em frente a uma plataforma de lançamento de foguetes de forma que o eixo da lente da sua câmera está a 50 m de distância e perfeitamente alinhada com a base do foguete. Em que ângulo ele deve colocar a câmera para tirar uma foto deste foguete quando ele estiver há 200 m de altura. 

Para conseguir os pontos desta atividade você deverá.
- (a) Encontrar o ângulo, usando a linguagem de programação Python;
- (b) Criar uma função em Python que ajude o fotógrafo a ajustar o ângulo da câmera para qualquer altitude
- (c) Plotar, usando a biblioteca Matplotlib um diagrama do triângulo resultante em cada altitude.

```python
# Exercício 1 - Letra (A)
alturaFoguete = 200
distanciaFoguete = 50

angulo_RD = math.atan(alturaFoguete/distanciaFoguete)
angulo_Graus = math.degrees(angulo_RD)

print(f"O angulo necesário é de: {angulo_Graus:.2f} graus")
```

```python
# Exercício 1 - Letra (B)
def calcAnguloFoto(p_alturaFoguete, p_distanciaFoguete):
    angulo_RD = math.atan(p_alturaFoguete/p_distanciaFoguete)
    return math.degrees(angulo_RD)

distanciaFoguete = float(input("Insira a distancia do foguete: "))
alturaFoguete = float(input("Insira a altura do foguete: "))
print(f"O angulo necesário é de: {calcAnguloFoto(alturaFoguete, distanciaFoguete):.2f} graus")
```

```python
# Exercício 1 - Letra (C)

distanciaFoguete = float(input("Insira a distancia do foguete: " ))
alturaFoguete = float(input("Insira a altura do foguete: " ))
angulo = calcAnguloFoto(alturaFoguete,distanciaFoguete)


x1=np.array([0,distanciaFoguete,distanciaFoguete,0]) 
y1=np.array([0,0, alturaFoguete,0]) 
fig1=plt.figure(1)
plt.plot(x1,y1,color=[0/255,100/255,200/255])


plt.grid(axis='both',which='major',color=(0.8, 0.8, 0.8, 0.5), linestyle='-', linewidth=1)
plt.minorticks_on()
plt.grid(axis='both',which='minor',color=(0.8, 0.8, 0.8, 0.5), linestyle=':', linewidth=1)

plt.xlabel("Distancia")
plt.ylabel("Altura")
plt.text(5,2,round(angulo,2),fontsize=15)
plt.show()
```

3. Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação e suas bibliotecas crie uma função que receba um ângulo, em graus, e plote um circulo trigonométrico de raio 1, centrado na origem destacando este ângulo, seu seno, seu cosseno e sua tangente.

```python
# Exercício 3

import matplotlib.pyplot as plt
import numpy as np
def plot(angulo):
    # Converter o ângulo para radianos
    angulo = np.deg2rad(angulo)
    # Definir as coordenadas do círculo trigonométrico
    theta = np.linspace(0, 2*np.pi, 1000)
    x = np.cos(theta)
    y = np.sin(theta)
    # Plotar o círculo trigonométrico
    fig, n1 = plt.subplots(figsize=(6, 6))
    n1.plot(x, y, color='black')
    n1.set_aspect('equal')
    n1.axhline(y=0, color='black', lw=1)
    n1.axvline(x=0, color='black', lw=1)
    # Destacar o ângulo no círculo
    n1.plot([0, np.cos(angulo)], [0, np.sin(angulo)], color='red')
    # Calcular e plotar o cosseno e o seno do ângulo
    n1.plot([np.cos(angulo), np.cos(angulo)], [0, np.sin(angulo)], linestyle='--', color='gray')
    n1.plot([0, np.cos(angulo)], [np.sin(angulo), np.sin(angulo)], linestyle='--', color='gray')
    n1.annotate(f'sen({angulo:.2f}) = {np.sin(angulo):.2f}', xy=(0.5*np.cos(angulo), 0.5*np.sin(angulo)), xytext=(0.7*np.cos(angulo), 0.7*np.sin(angulo)))
    n1.annotate(f'cos({angulo:.2f}) = {np.cos(angulo):.2f}', xy=(0.5*np.cos(angulo), 0.5*np.sin(angulo)), xytext=(0.4*np.cos(angulo), 0.4*np.sin(angulo)))
    # Calcular e plotar a tangente do ângulo
    if np.cos(angulo) != 0:
        tan_angle = np.tan(angulo)
        x_tan = np.linspace(-1, 1, 100)
        y_tan = tan_angle*x_tan
        n1.plot(x_tan, y_tan, linestyle='--', color='gray')
        n1.annotate(f'tan({angulo:.2f}) = {tan_angle:.2f}', xy=(0.5, 0.5*tan_angle), xytext=(0.7, 0.7*tan_angle))
        plt.show()
    else:
        n1.annotate('tan não definido', xy=(0.5, 0.5), xytext=(0.5, 0.5))
        plt.show()

# Chamar a função
plot(int(input("Digite o ângulo (em graus): ")))
```

4. Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação e suas bibliotecas crie uma função que receba um número inteiro tal que 
 e plote as funções seno, cosseno e tangente para todos os ângulos possíveis considerando que  representa o número períodos completos.

```python
# Exercício 4

def plotFuncTrig(periodos):
    x = np.linspace(0, 2*np.pi*periodos, 1000)
    ySeno = np.sin(x)
    yCosseno = np.cos(x)
    yTangente = np.tan(x)

    fig, axs = plt.subplots(3, 1, figsize=(8, 12), sharex=True)
    # Gráfico do Seno
    axs[0].plot(x, ySeno, label='Seno')
    axs[0].legend()
    # Gráfico do Cosseno
    axs[1].plot(x, yCosseno, label='Cosseno')
    axs[1].legend()
    # Gráfico da Tangente
    axs[2].plot(x, yTangente, label='Tangente')
    axs[2].legend()

    axs[0].set_ylabel('Seno(x)')
    axs[1].set_ylabel('Cosseno(x)')
    axs[2].set_ylabel('Tangente(x)')
    axs[2].set_xlabel('X')

    plt.show()

plotFuncTrig(int(input("Digite número períodos: ")))
```
